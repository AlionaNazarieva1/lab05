```python
import numpy as np
import math
from scipy.integrate import odeint
import matplotlib.pyplot as plt
a= 0.25
b= 0.6
c= 0.05
d= 0.061
t0 = 0
tmax = 400
dt = 0.1
def dy(s,t):
    dy1 =  -a*s[0] + c*s[0]*s[1]
    dy2 =  b*s[1] - d*s[0]*s[1]
    return [dy1, dy2]
t = np.arange(t0,tmax,dt) 
v0=[b/d,a/c]
s = odeint(dy,v0,t)
v0
```

```python
plt.plot(s[:,0],s[:,1],'r--', linewidth=2.0,label="фазовый портрет")
plt.legend()
plt.grid()
plt.show()

plt.plot(t, s[:,0],'r--', linewidth=2.0,label=" изменения числа популяции хищников ")
plt.legend()
plt.grid()
plt.show()

plt.plot(t, s[:,1],'r--', linewidth=2.0,label=" изменения числа популяции жертв ")
plt.legend()
plt.grid()
plt.show()
```
